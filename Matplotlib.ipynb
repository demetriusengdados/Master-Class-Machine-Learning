```
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d
import numpy as np
```

```
#Retas no plano
plt.plot([1, 2, 6, 4])
```

```
#Estilizando o gráfico
plt.plot([10, 20, 30, 40],[1, 5, 9, 8])
plt.ylabel('Alunos')
plt.xlabel('Notas')
plt.title('Relação entre id do alunos e notas')
plt.show()
```

```
#Gráfico de barras
plt.plot([10, 20, 30, 40],[1, 5, 9, 8], label='Sala 1')
plt.plot([10, 25, 35, 45],[1, 15, 9, 12], label='Sala 2')
plt.ylabel('Alunos')
plt.xlabel('Notas')
plt.title('Relação entre id do alunos e notas')
plt.legend()
plt.show()
```

```
x = [1,3,5,7]
y = [2,3,6,10]

x2 = [2,4,6,8]
y2 = [5,6,7,9]

plt.bar(x, y, color = 'blue', label='label 1')
plt.bar(x2, y2, color = 'green', label='label 2')
plt.ylabel('Altura média')
plt.legend()
plt.xlabel('Xlabel')
plt.title('Um titulo')
plt.show()
```

```
#Gráfico de dispersão
xp = [1, 1, 1.5, 2, 2]
yp = [3, 3.5, 4, 4.5, 5]
xp1 = [3, 3, 3.5, 4, 2]
yp1 = [2, 2.5, 3, 3.5, 2]
x = [1, 2, 3, 4]
y = [2, 3, 4, 5]
plt.plot(x,y)
plt.scatter(xp, yp)
plt.scatter(xp1, yp1)
```

```
#Grafico de Pizza
label = ['Floresta', 'Deserto']
dados = [10, 60]

fig, ax = plt.subplots()

ax.pie(dados, labels=label, autopct = '%1.1f%%')
plt.show()
```
